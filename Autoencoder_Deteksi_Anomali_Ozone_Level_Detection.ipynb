```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Kode Program: Autoencoder Deteksi Anomali (Ozone Level Detection, UCI ID=172)
# Dataset : https://archive.ics.uci.edu/dataset/172/ozone+level+detection
# Repo    : https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingk.  : Google Colab
# Tujuan  : Melatih autoencoder dari data normal, mendeteksi anomali via reconstruction error,
#           validasi dengan label (bila tersedia), serta visualisasi + tabel audit.
# ==========================================================

# =========================
# 0) SETUP
# =========================
!pip -q install ucimlrepo tensorflow scikit-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    roc_auc_score, roc_curve
)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["font.size"] = 11

RANDOM_STATE = 42

# =========================
# 1) AKUISISI DATA (UCI)
# =========================
dataset_id = 172  # Ozone Level Detection
ds = fetch_ucirepo(id=dataset_id)

X_raw = ds.data.features.copy()
y_raw = ds.data.targets.copy() if ds.data.targets is not None else None

print("Ukuran fitur (X):", X_raw.shape)
display(X_raw.head(3))

if y_raw is not None:
    print("Ukuran target (y):", y_raw.shape)
    display(y_raw.head(3))

# =========================
# 2) PRAPROSES (NUMERIK + IMPUTASI + LABEL BERSIH)
# =========================
# 2.1 Ambil fitur numerik saja agar kompatibel untuk dense autoencoder
X = X_raw.select_dtypes(include=[np.number]).copy()
X = X.replace([np.inf, -np.inf], np.nan)

# 2.2 Imputasi median: aman untuk data sensor/cuaca dan tidak membuat pola semu seperti interpolasi
X = X.fillna(X.median(numeric_only=True))

print("Ukuran setelah seleksi numerik + imputasi:", X.shape)

# 2.3 Rapikan label bila ada, jadikan vektor 0/1
y = None
if y_raw is not None:
    if isinstance(y_raw, pd.DataFrame):
        y = y_raw.iloc[:, 0].values
    else:
        y = np.array(y_raw).ravel()

    # Konversi ke int jika mungkin; jika tidak, map unik ke 0/1
    try:
        y = y.astype(int)
    except:
        uniq = pd.Series(y).unique().tolist()
        if len(uniq) >= 2:
            mapping = {uniq[0]: 0, uniq[1]: 1}
        else:
            mapping = {uniq[0]: 0}
        y = pd.Series(y).map(mapping).fillna(0).astype(int).values

    print("Distribusi label (seluruh data):", pd.Series(y).value_counts().to_dict())

# =========================
# 3) SPLIT DATA TANPA DATA LEAKAGE
# =========================
# Logika utama:
# - Jika label ada, autoencoder dilatih dari data normal (y=0) agar anomali tidak "dipelajari" sebagai normal.
# - Split dilakukan terlebih dahulu supaya evaluasi test tetap adil.
if y is not None:
    X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(
        X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
    )

    X_train_raw = X_train_all[y_train_all == 0].copy()  # train hanya normal
    X_test_raw  = X_test_all.copy()                     # test campuran

    print("Train (normal saja):", X_train_raw.shape, "| Test (campuran):", X_test_raw.shape)
    print("Distribusi label (test):", pd.Series(y_test_all).value_counts().to_dict())
else:
    X_train_raw, X_test_raw = train_test_split(
        X, test_size=0.25, random_state=RANDOM_STATE
    )
    print("Train:", X_train_raw.shape, "| Test:", X_test_raw.shape)

# Scaling hanya dari train (mencegah leakage)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_test  = scaler.transform(X_test_raw)

# =========================
# 4) BANGUN AUTOENCODER (ENCODER–BOTTLENECK–DECODER)
# =========================
input_dim = X_train.shape[1]
bottleneck = max(4, input_dim // 3)

inputs = keras.Input(shape=(input_dim,))
h = layers.Dense(bottleneck * 2, activation="relu")(inputs)
h = layers.Dense(bottleneck, activation="relu")(h)         # bottleneck
h = layers.Dense(bottleneck * 2, activation="relu")(h)
outputs = layers.Dense(input_dim, activation="linear")(h)  # output kontinu

autoencoder = keras.Model(inputs, outputs)
autoencoder.compile(optimizer=keras.optimizers.Adam(1e-3), loss="mse")

autoencoder.summary()

# =========================
# 5) TRAINING (EARLY STOPPING UNTUK CEGAH OVERFITTING)
# =========================
callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True)
]

history = autoencoder.fit(
    X_train, X_train,
    epochs=60,
    batch_size=64,
    validation_split=0.2,
    shuffle=True,
    callbacks=callbacks,
    verbose=1
)

# Gambar 1: Kurva training vs validation
plt.figure()
plt.plot(history.history["loss"], label="loss (train)")
plt.plot(history.history["val_loss"], label="loss (val)")
plt.title("Gambar 1. Kurva Pelatihan Autoencoder (MSE)")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.legend()
plt.show()

# =========================
# 6) SKOR ANOMALI: RECONSTRUCTION ERROR + THRESHOLD KUANTIL
# =========================
X_train_pred = autoencoder.predict(X_train, verbose=0)
X_test_pred  = autoencoder.predict(X_test,  verbose=0)

train_err = np.mean(np.square(X_train - X_train_pred), axis=1)
test_err  = np.mean(np.square(X_test  - X_test_pred),  axis=1)

# Threshold = persentil 95 dari train error (train berisi normal), sesuai logika deteksi anomali
threshold = np.percentile(train_err, 95)
test_pred_anom = (test_err > threshold).astype(int)

print("\nRingkasan reconstruction error:")
print(f"Train mean={train_err.mean():.6f}, std={train_err.std():.6f}")
print(f"Test  mean={test_err.mean():.6f}, std={test_err.std():.6f}")
print(f"Threshold p95(train) = {threshold:.6f}")
print(f"Proporsi prediksi anomali pada test = {test_pred_anom.mean()*100:.2f}%")

# Gambar 2: Histogram error + threshold
plt.figure()
plt.hist(train_err, bins=40, alpha=0.7, label="Train error (normal)")
plt.hist(test_err,  bins=40, alpha=0.7, label="Test error")
plt.axvline(threshold, linestyle="--", label="Threshold p95")
plt.title("Gambar 2. Distribusi Reconstruction Error dan Ambang Anomali")
plt.xlabel("Reconstruction error (MSE)")
plt.ylabel("Frekuensi")
plt.legend()
plt.show()

# Gambar 3: Error test berurutan + titik anomali
plt.figure()
plt.plot(test_err, label="Test error")
plt.axhline(threshold, linestyle="--", label="Threshold p95")
anom_idx = np.where(test_pred_anom == 1)[0]
plt.scatter(anom_idx, test_err[anom_idx], s=18, label="Prediksi anomali")
plt.title("Gambar 3. Deret Reconstruction Error pada Data Uji")
plt.xlabel("Index baris (test)")
plt.ylabel("Reconstruction error (MSE)")
plt.legend()
plt.show()

# =========================
# 7) TABEL AUDIT: TOP-15 ERROR TERBESAR
# =========================
audit = X_test_raw.copy()
audit["reconstruction_error"] = test_err
audit["pred_is_anomaly"] = test_pred_anom

if y is not None:
    audit["true_label"] = y_test_all

top15 = audit.sort_values("reconstruction_error", ascending=False).head(15)
print("\nTabel 1. 15 Sampel dengan Reconstruction Error Tertinggi (Data Uji)")
display(top15)

# =========================
# 8) EVALUASI DENGAN LABEL (JIKA TERSEDIA)
# =========================
if y is not None:
    print("\nEvaluasi berbasis label:")
    print(classification_report(y_test_all, test_pred_anom, digits=4))

    cm = confusion_matrix(y_test_all, test_pred_anom)
    ConfusionMatrixDisplay(confusion_matrix=cm).plot(values_format="d")
    plt.title("Gambar 4. Confusion Matrix Deteksi Anomali (Threshold p95)")
    plt.show()

    # ROC/AUC memakai skor kontinu test_err (bukan prediksi kelas)
    auc = roc_auc_score(y_test_all, test_err)
    fpr, tpr, _ = roc_curve(y_test_all, test_err)

    plt.figure()
    plt.plot(fpr, tpr, label=f"AUC = {auc:.4f}")
    plt.title("Gambar 5. ROC Curve (Skor = Reconstruction Error)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

# =========================
# 9) VISUAL ASLI vs REKONSTRUKSI (CONTOH 3 FITUR)
# =========================
# Tujuan logis:
# - Menunjukkan kemampuan autoencoder merekonstruksi beberapa fitur atmosfer sekaligus
# - Anomali biasanya terlihat sebagai deviasi konsisten pada lebih dari satu fitur
# - Visual multifikur memperkuat interpretasi lintas-variabel

# Pilih 3 fitur pertama (dapat diganti sesuai kebutuhan analisis)
feature_names = X_train_raw.columns[:3]
seg_len = min(200, len(X_test_raw))

fig, axes = plt.subplots(3, 1, figsize=(10, 9), sharex=True)

for i, feat in enumerate(feature_names):
    true_seg = X_test[:seg_len, i]
    reco_seg = X_test_pred[:seg_len, i]

    axes[i].plot(true_seg, label="Asli (scaled)")
    axes[i].plot(reco_seg, label="Rekonstruksi (scaled)", alpha=0.85)
    axes[i].set_title(f"Fitur {i+1}: {feat}")
    axes[i].set_ylabel("Nilai (skala standar)")
    axes[i].legend(loc="upper right")

axes[-1].set_xlabel("Index segmen (data uji)")

plt.suptitle(
    "Gambar 6. Perbandingan Nilai Asli vs Rekonstruksi pada Tiga Fitur Atmosfer",
    fontsize=13
)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

```
