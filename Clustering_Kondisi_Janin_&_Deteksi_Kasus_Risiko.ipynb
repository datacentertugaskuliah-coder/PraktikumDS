```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Proyek : Cardiotocography â€“ Clustering Kondisi Janin & Deteksi Kasus Risiko (DBSCAN)
# Dataset: Cardiotocography (UCI ID=193) - CTG.xls
# Sumber : UCI Machine Learning Repository
# Repo   : https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingk. : Google Colab
# ==========================================================

# 0) IMPORT LIBRARY
# ----------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
from sklearn.metrics import (
    adjusted_rand_score,
    homogeneity_score,
    completeness_score,
    v_measure_score,
    silhouette_score
)

plt.rcParams["figure.figsize"] = (7, 5)
plt.rcParams["font.size"] = 11

# 1) LOAD & CLEANING DATASET CTG DARI UCI
# ----------------------------------------------------------
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00193/CTG.xls"

# Sheet "Data" berisi 2126 baris CTG, beberapa baris akhir adalah footer
raw_df = pd.read_excel(url, sheet_name="Data", skipfooter=3)

# Hapus kolom-kolom 'Unnamed' (biasanya kosong atau header ganda)
df = raw_df.drop(columns=raw_df.filter(like="Unnamed").columns)

# Baris pertama berisi nama variabel -> jadikan header
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)

# Konversi seluruh kolom ke numerik (float)
df = df.astype(float)

print("=== INFORMASI DATA AWAL ===")
print("Ukuran data (baris, kolom):", df.shape)
print("\nBeberapa baris awal:")
display(df.head())

# 2) EXPLORATORY DATA ANALYSIS (EDA) RINGAN
# ----------------------------------------------------------
print("\n=== RINGKASAN STATISTIK FITUR NUMERIK ===")
display(df.describe().T)

# Label NSP: 1=Normal, 2=Suspect, 3=Pathologic
nsp_map = {1: "Normal", 2: "Suspect", 3: "Pathologic"}
df["NSP"] = df["NSP"].astype(int)
df["NSP_label"] = df["NSP"].map(nsp_map)

print("\n=== DISTRIBUSI KELAS NSP ===")
print(df["NSP_label"].value_counts())

# Gambar 1: Distribusi kelas NSP
plt.figure()
df["NSP_label"].value_counts().plot(kind="bar")
plt.title("Gambar 1. Distribusi Kelas Fetal State (NSP)")
plt.xlabel("Kelas Fetal State")
plt.ylabel("Jumlah Sampel")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# 3) PREPROCESSING: PILIH FITUR + STANDARDISASI
# ----------------------------------------------------------
# 21 fitur fisiologis (LB sampai Tendency) sebagai input unsupervised
feature_cols = [
    "LB","AC","FM","UC","DL","DS","DP",
    "ASTV","MSTV","ALTV","MLTV",
    "Width","Min","Max","Nmax","Nzeros",
    "Mode","Mean","Median","Variance","Tendency"
]

X = df[feature_cols].values
y_nsp = df["NSP"].values  # 1=Normal, 2=Suspect, 3=Pathologic

# Standardisasi fitur agar tiap fitur punya skala sebanding
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\n=== INFORMASI FITUR ===")
print("Bentuk fitur (setelah scaling):", X_scaled.shape)

# 4) EKSPLORASI PARAMETER DBSCAN (eps & min_samples)
# ----------------------------------------------------------
eps_candidates = [0.5, 1.0, 1.5, 2.0]
min_samples_candidates = [5, 10, 20]

results = []

for eps in eps_candidates:
    for min_s in min_samples_candidates:
        dbs_tmp = DBSCAN(eps=eps, min_samples=min_s, metric="euclidean")
        labels_tmp = dbs_tmp.fit_predict(X_scaled)
        n_clusters = len(set(labels_tmp)) - (1 if -1 in labels_tmp else 0)
        n_noise = list(labels_tmp).count(-1)
        noise_ratio = n_noise / len(labels_tmp)
        results.append((eps, min_s, n_clusters, noise_ratio))

# Tampilkan hasil eksplorasi dalam bentuk tabel
res_df = pd.DataFrame(
    results,
    columns=["eps", "min_samples", "n_clusters", "noise_ratio"]
)
res_df["noise_ratio(%)"] = res_df["noise_ratio"] * 100

print("\n=== EKSPLORASI PARAMETER DBSCAN (SEMUA KOMBINASI) ===")
display(res_df.sort_values(["eps", "min_samples"]).reset_index(drop=True))

# Logika pemilihan parameter yang lebih logis:
# - Target jumlah cluster â‰ˆ 3 (mencerminkan Normal, Suspect, Pathologic)
# - Noise moderat sekitar 20% (tidak terlalu sedikit, tidak terlalu banyak)
TARGET_CLUSTERS = 3
TARGET_NOISE = 0.20

candidates = res_df[
    (res_df["n_clusters"] >= 2) &
    (res_df["n_clusters"] <= 6) &
    (res_df["noise_ratio"] >= 0.05) &
    (res_df["noise_ratio"] <= 0.40)
].copy()

if len(candidates) > 0:
    candidates["cluster_distance"] = (candidates["n_clusters"] - TARGET_CLUSTERS).abs()
    candidates["noise_distance"] = (candidates["noise_ratio"] - TARGET_NOISE).abs()
    candidates = candidates.sort_values(
        ["cluster_distance", "noise_distance", "eps", "min_samples"]
    )

    best_row = candidates.iloc[0]
    final_eps = float(best_row["eps"])
    final_min_samples = int(best_row["min_samples"])

    print("\n=== PARAMETER TERPILIH (BERDASARKAN KRITERIA LOGIS) ===")
    display(candidates.head())
else:
    # Fallback jika tidak ada kandidat yang memenuhi kriteria sederhana
    final_eps = 1.5
    final_min_samples = 10
    print("\nTidak ada kandidat yang memenuhi kriteria sederhana.")
    print("Gunakan nilai default yang dianggap cukup seimbang.")

print(f"\nParameter DBSCAN final: eps = {final_eps}, min_samples = {final_min_samples}")

# 5) TRAINING DBSCAN FINAL
# ----------------------------------------------------------
dbscan = DBSCAN(
    eps=final_eps,
    min_samples=final_min_samples,
    metric="euclidean"
)
cluster_labels = dbscan.fit_predict(X_scaled)

df["cluster_dbscan"] = cluster_labels

n_clusters_ = len(set(cluster_labels)) - (1 if -1 in cluster_labels else 0)
n_noise_ = list(cluster_labels).count(-1)

print("\n=== HASIL CLUSTERING DBSCAN (FINAL) ===")
print(f"Jumlah cluster (tidak termasuk noise): {n_clusters_}")
print(f"Jumlah noise points: {n_noise_} "
      f"({n_noise_/len(cluster_labels)*100:.1f}%)")

# 6) EVALUASI CLUSTERING
# ----------------------------------------------------------
# 6a) Evaluasi eksternal terhadap label NSP (Normal/Suspect/Pathologic)
ari = adjusted_rand_score(y_nsp, cluster_labels)
homo = homogeneity_score(y_nsp, cluster_labels)
comp = completeness_score(y_nsp, cluster_labels)
v_meas = v_measure_score(y_nsp, cluster_labels)

print("\n=== EVALUASI CLUSTERING vs LABEL NSP ===")
print(f"Adjusted Rand Index (ARI) : {ari:.3f}")
print(f"Homogeneity               : {homo:.3f}")
print(f"Completeness              : {comp:.3f}")
print(f"V-measure                 : {v_meas:.3f}")

# 6b) Evaluasi internal: silhouette score (tanpa noise)
unique_clusters = set(cluster_labels)
unique_clusters_no_noise = [c for c in unique_clusters if c != -1]

if len(unique_clusters_no_noise) >= 2:
    mask_non_noise = cluster_labels != -1
    try:
        sil = silhouette_score(X_scaled[mask_non_noise],
                               cluster_labels[mask_non_noise])
        print(f"Silhouette score (tanpa noise): {sil:.3f}")
    except Exception as e:
        print("Silhouette score tidak dapat dihitung:", e)
else:
    print("Silhouette score tidak dihitung karena hanya 1 cluster (plus noise).")

# 6c) Tabel silang cluster vs kelas NSP
ct = pd.crosstab(df["NSP_label"], df["cluster_dbscan"])
print("\n=== TABEL SILANG NSP vs CLUSTER DBSCAN ===")
display(ct)

# 7) VISUALISASI PCA 2D (NSP vs CLUSTER)
# ----------------------------------------------------------
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_scaled)

df["PC1"] = X_2d[:, 0]
df["PC2"] = X_2d[:, 1]

# Gambar 2: PCA 2D berwarna berdasarkan kelas NSP
plt.figure()
for label, name in nsp_map.items():
    mask = df["NSP"] == label
    plt.scatter(df.loc[mask, "PC1"], df.loc[mask, "PC2"],
                label=name, alpha=0.7, s=20)

plt.title("Gambar 2. Visualisasi PCA 2D Berdasarkan Kelas NSP")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.tight_layout()
plt.show()

# Gambar 3: PCA 2D berwarna berdasarkan cluster DBSCAN
plt.figure()
unique_clusters_sorted = sorted(set(cluster_labels))

for cl in unique_clusters_sorted:
    mask = df["cluster_dbscan"] == cl
    label_name = f"Cluster {cl}" if cl != -1 else "Noise (-1)"
    plt.scatter(df.loc[mask, "PC1"], df.loc[mask, "PC2"],
                label=label_name, alpha=0.7, s=20)

plt.title("Gambar 3. Visualisasi PCA 2D Berdasarkan Cluster DBSCAN")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.tight_layout()
plt.show()

# 8) HEATMAP: NSP vs CLUSTER DBSCAN
# ----------------------------------------------------------
ct_values = ct.values

plt.figure(figsize=(6, 4))
plt.imshow(ct_values, aspect="auto")
plt.colorbar(label="Jumlah Sampel")
plt.xticks(ticks=range(ct.shape[1]), labels=ct.columns, rotation=0)
plt.yticks(ticks=range(ct.shape[0]), labels=ct.index)
plt.xlabel("Cluster DBSCAN")
plt.ylabel("Kelas NSP")
plt.title("Gambar 4. Heatmap NSP vs Cluster DBSCAN")
plt.tight_layout()
plt.show()

# 9) RINGKASAN INTERPRETASI (SINGKAT)
# ----------------------------------------------------------
print("\n=== RINGKASAN INTERPRETASI (SINGKAT) ===")
print("- DBSCAN membentuk cluster berdasarkan kepadatan titik di ruang fitur CTG.")
print("- Noise (-1) merepresentasikan rekaman CTG yang tidak cocok dengan pola mayoritas.")
print("- Cluster yang didominasi kelas Suspect/Pathologic pada tabel silang & heatmap")
print("  dapat diinterpretasikan sebagai kelompok CTG berisiko.")
print("- Visualisasi PCA 2D membantu melihat apakah area dengan dominasi Pathologic")
print("  beririsan dengan cluster tertentu, sehingga relevan untuk pemantauan klinis.")
```
