```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Pembahasan:  NLP Text Preprocessing, Tokenisasi, dan Klasifikasi Sentimen (Biner)
# Dataset: Sentiment Labelled Sentences (UCI Machine Learning Repository, ID: 331)
# Sumber: https://archive.ics.uci.edu/dataset/331/sentiment+labelled+sentences
# Repo: https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingkungan: Google Colab
# ==========================================================

# ==========================================================
# 0) SETUP ENVIRONMENT
# ==========================================================
!pip -q install pandas numpy scikit-learn matplotlib wordcloud

import os, io, zipfile, re, random
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support,
    confusion_matrix, classification_report,
    roc_curve, auc
)

from wordcloud import WordCloud
import joblib

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)

pd.set_option("display.max_colwidth", 140)
plt.rcParams["figure.figsize"] = (8, 5)

# ==========================================================
# 1) DOWNLOAD + EKSTRAK DATA (UCI)
# ==========================================================
DATA_URL = "https://archive.ics.uci.edu/static/public/331/sentiment%2Blabelled%2Bsentences.zip"
OUT_DIR = "uci_331_sentiment_labelled_sentences"
os.makedirs(OUT_DIR, exist_ok=True)

resp = requests.get(DATA_URL, timeout=60)
resp.raise_for_status()

with zipfile.ZipFile(io.BytesIO(resp.content)) as z:
    z.extractall(OUT_DIR)

# Kumpulkan kandidat file .txt
all_txt = []
for root, _, files in os.walk(OUT_DIR):
    for f in files:
        if f.lower().endswith(".txt"):
            all_txt.append(os.path.join(root, f))

# Fokus ke 3 file utama dataset
EXPECTED = {"amazon_cells_labelled.txt", "imdb_labelled.txt", "yelp_labelled.txt"}
txt_files = [p for p in all_txt if os.path.basename(p) in EXPECTED]

# Fallback bila struktur folder berbeda
if len(txt_files) == 0:
    txt_files = [p for p in all_txt if "label" in os.path.basename(p).lower()]

print("File yang dipakai:")
for p in txt_files:
    print("-", os.path.basename(p))

# ==========================================================
# 2) LOAD DATA (ANTI UnicodeDecodeError)
# ==========================================================
def infer_source(path: str) -> str:
    base = os.path.basename(path).lower()
    if "amazon" in base:
        return "amazon"
    if "imdb" in base:
        return "imdb"
    if "yelp" in base:
        return "yelp"
    return "unknown"

def safe_decode_bytes(raw: bytes) -> str:
    for enc in ("utf-8", "utf-8-sig", "cp1252", "latin-1"):
        try:
            return raw.decode(enc)
        except UnicodeDecodeError:
            continue
    return raw.decode("utf-8", errors="replace")

def load_one(path: str) -> pd.DataFrame:
    raw = open(path, "rb").read()
    text = safe_decode_bytes(raw)

    df_ = pd.read_csv(
        io.StringIO(text),
        sep="\t",
        header=None,
        names=["sentence", "label"],
        engine="python",
        on_bad_lines="skip"
    )
    df_["source"] = infer_source(path)
    return df_

df = pd.concat([load_one(p) for p in txt_files], ignore_index=True)

# Standarisasi tipe & bersihkan label
df["sentence"] = df["sentence"].astype(str)
df["label"] = pd.to_numeric(df["label"], errors="coerce")
df = df.dropna(subset=["label"]).copy()
df["label"] = df["label"].astype(int)

# Jaga hanya label biner 0/1
df = df[df["label"].isin([0, 1])].copy()

print("Ukuran data:", df.shape)
display(df.head(10))

# ==========================================================
# 3) DATA QUALITY CHECK
# ==========================================================
print("\nDistribusi label (overall):")
display(df["label"].value_counts().sort_index())

print("\nDistribusi label per sumber:")
display(pd.crosstab(df["source"], df["label"], margins=True))

dup_count = df.duplicated(subset=["sentence", "label", "source"]).sum()
print("\nDuplikasi baris (sentence+label+source):", dup_count)

# ==========================================================
# 4) EDA + VISUALISASI (TABEL & GRAFIK)
# ==========================================================
def clean_for_stats(text: str) -> str:
    text = text.lower()
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"[^a-z0-9\s']", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_tmp"] = df["sentence"].map(clean_for_stats)
df["char_len"] = df["clean_tmp"].str.len()
df["token_len"] = df["clean_tmp"].apply(lambda s: 0 if not s else len(s.split()))

summary_stats = df.groupby("label")[["char_len", "token_len"]].agg(["count", "mean", "median", "min", "max"])
print("\nRingkasan statistik panjang teks (per label):")
display(summary_stats)

# Grafik distribusi label
counts = df["label"].value_counts().sort_index()
plt.figure()
plt.bar(counts.index.astype(str), counts.values)
plt.title("Distribusi Label Sentimen (0=Negatif, 1=Positif)")
plt.xlabel("Label")
plt.ylabel("Jumlah")
plt.show()

# Grafik distribusi label per sumber
ct = pd.crosstab(df["source"], df["label"])
x = np.arange(len(ct.index))
w = 0.35
plt.figure()
plt.bar(x - w/2, ct[0].values, width=w, label="0 (Negatif)")
plt.bar(x + w/2, ct[1].values, width=w, label="1 (Positif)")
plt.xticks(x, ct.index)
plt.title("Distribusi Label per Sumber")
plt.xlabel("Sumber")
plt.ylabel("Jumlah")
plt.legend()
plt.show()

# Histogram panjang kalimat
plt.figure()
plt.hist(df["token_len"], bins=30)
plt.title("Distribusi Panjang Kalimat (Jumlah Token)")
plt.xlabel("Token per kalimat")
plt.ylabel("Frekuensi")
plt.show()

# ==========================================================
# 5) SPLIT TRAINTEST (STRATIFIED)
# ==========================================================
X = df["sentence"].values
y = df["label"].values

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Ukuran train:", len(X_train), "| Ukuran test:", len(X_test))

# ==========================================================
# 6) BASELINE: TF-IDF + LOGISTIC REGRESSION
# ==========================================================
baseline_lr = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True,
        strip_accents="unicode",
        token_pattern=r"(?u)\b\w[\w']+\b",
        ngram_range=(1, 1),
        min_df=1,
        max_df=0.95
    )),
    ("clf", LogisticRegression(max_iter=2000, random_state=RANDOM_STATE))
])

baseline_lr.fit(X_train, y_train)
pred_base = baseline_lr.predict(X_test)
proba_base = baseline_lr.predict_proba(X_test)[:, 1]

acc = accuracy_score(y_test, pred_base)
p, r, f1, _ = precision_recall_fscore_support(y_test, pred_base, average="binary", zero_division=0)

print("\n=== BASELINE (TF-IDF + LogisticRegression) ===")
print(f"Accuracy : {acc:.4f}")
print(f"Precision: {p:.4f}")
print(f"Recall   : {r:.4f}")
print(f"F1-score : {f1:.4f}")
print("\nClassification report:\n", classification_report(y_test, pred_base, digits=4))

# Confusion Matrix baseline
cm = confusion_matrix(y_test, pred_base)
plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix (Baseline)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks([0, 1], ["0", "1"])
plt.yticks([0, 1], ["0", "1"])
for (i, j), v in np.ndenumerate(cm):
    plt.text(j, i, str(v), ha="center", va="center")
plt.colorbar()
plt.show()

# ROC baseline
fpr, tpr, _ = roc_curve(y_test, proba_base)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.4f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("ROC Curve (Baseline LR)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

# ==========================================================
# 7) TUNING: NGRAM + REGULARISASI (GridSearchCV)
# ==========================================================
param_grid = {
    "tfidf__ngram_range": [(1, 1), (1, 2)],
    "tfidf__min_df": [1, 2, 3],
    "tfidf__max_df": [0.90, 0.95, 1.0],
    "clf__C": [0.3, 1.0, 3.0]
}

grid = GridSearchCV(
    estimator=baseline_lr,
    param_grid=param_grid,
    scoring="f1",
    cv=5,
    n_jobs=-1,
    verbose=0
)

grid.fit(X_train, y_train)
best_lr = grid.best_estimator_

print("\nBest params (Tuned LR):", grid.best_params_)
print("Best CV F1 (Tuned LR):", grid.best_score_)

pred_lr = best_lr.predict(X_test)
proba_lr = best_lr.predict_proba(X_test)[:, 1]

acc_lr = accuracy_score(y_test, pred_lr)
p_lr, r_lr, f1_lr, _ = precision_recall_fscore_support(y_test, pred_lr, average="binary", zero_division=0)

print("\n=== TUNED LR (Test) ===")
print(f"Accuracy : {acc_lr:.4f}")
print(f"Precision: {p_lr:.4f}")
print(f"Recall   : {r_lr:.4f}")
print(f"F1-score : {f1_lr:.4f}")

# ==========================================================
# 8) MODEL PEMBANDING: TF-IDF + LinearSVC
# ==========================================================
svc_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True,
        strip_accents="unicode",
        token_pattern=r"(?u)\b\w[\w']+\b",
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.95
    )),
    ("clf", LinearSVC(random_state=RANDOM_STATE))
])

svc_pipe.fit(X_train, y_train)
pred_svc = svc_pipe.predict(X_test)

acc_svc = accuracy_score(y_test, pred_svc)
p_svc, r_svc, f1_svc, _ = precision_recall_fscore_support(y_test, pred_svc, average="binary", zero_division=0)

print("\n=== LinearSVC (Test) ===")
print(f"Accuracy : {acc_svc:.4f}")
print(f"Precision: {p_svc:.4f}")
print(f"Recall   : {r_svc:.4f}")
print(f"F1-score : {f1_svc:.4f}")

# Tabel perbandingan
results = pd.DataFrame([
    {"Model": "Baseline TF-IDF + LogisticRegression", "Accuracy": acc, "Precision": p, "Recall": r, "F1": f1},
    {"Model": "Tuned TF-IDF + LogisticRegression",    "Accuracy": acc_lr, "Precision": p_lr, "Recall": r_lr, "F1": f1_lr},
    {"Model": "TF-IDF + LinearSVC",                   "Accuracy": acc_svc, "Precision": p_svc, "Recall": r_svc, "F1": f1_svc},
]).sort_values("F1", ascending=False)

print("\nRingkasan perbandingan model (urut F1 tertinggi):")
display(results)

# ==========================================================
# 9) PILIH MODEL FINAL (BERDASARKAN F1) + EVALUASI
# ==========================================================
best_name = results.iloc[0]["Model"]
if best_name == "TF-IDF + LinearSVC":
    final_model = svc_pipe
else:
    final_model = best_lr if best_name == "Tuned TF-IDF + LogisticRegression" else baseline_lr

print("Model final dipilih:", best_name)

pred_final = final_model.predict(X_test)
acc_f = accuracy_score(y_test, pred_final)
p_f, r_f, f1_f, _ = precision_recall_fscore_support(y_test, pred_final, average="binary", zero_division=0)

print("\n=== FINAL MODEL (Test) ===")
print(f"Accuracy : {acc_f:.4f}")
print(f"Precision: {p_f:.4f}")
print(f"Recall   : {r_f:.4f}")
print(f"F1-score : {f1_f:.4f}")
print("\nClassification report:\n", classification_report(y_test, pred_final, digits=4))

# Confusion Matrix final
cm_f = confusion_matrix(y_test, pred_final)
plt.figure()
plt.imshow(cm_f)
plt.title("Confusion Matrix (Final Model)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks([0, 1], ["0", "1"])
plt.yticks([0, 1], ["0", "1"])
for (i, j), v in np.ndenumerate(cm_f):
    plt.text(j, i, str(v), ha="center", va="center")
plt.colorbar()
plt.show()

# ROC final: probabilitas (LR) atau decision score (LinearSVC)
if hasattr(final_model.named_steps["clf"], "predict_proba"):
    scores = final_model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, scores)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.4f}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.title("ROC Curve (Final Model)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()
elif hasattr(final_model.named_steps["clf"], "decision_function"):
    scores = final_model.decision_function(X_test)
    fpr, tpr, _ = roc_curve(y_test, scores)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.4f}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.title("ROC Curve (Final Model, Decision Scores)")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()

# ==========================================================
# 10) ERROR ANALYSIS (CONTOH SALAH KLASIFIKASI)
# ==========================================================
errors = []
for txt, yt, yp in zip(X_test, y_test, pred_final):
    if yt != yp:
        errors.append({"sentence": txt, "y_true": yt, "y_pred": yp})

err_df = pd.DataFrame(errors)
print("Jumlah salah klasifikasi:", len(err_df))
display(err_df.head(12))

# ==========================================================
# 11) WORDCLOUD (POSITIF vs NEGATIF)
# ==========================================================
def make_wordcloud(texts, title):
    blob = " ".join([clean_for_stats(t) for t in texts])
    wc = WordCloud(width=900, height=450, background_color="white").generate(blob)
    plt.figure(figsize=(12, 5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(title)
    plt.show()

make_wordcloud(df.loc[df["label"] == 1, "sentence"].values, "WordCloud - Positif (Label 1)")
make_wordcloud(df.loc[df["label"] == 0, "sentence"].values, "WordCloud - Negatif (Label 0)")

# ==========================================================
# 12) INTERPRETASI FITUR (KHUSUS Logistic Regression)
# ==========================================================
if isinstance(final_model.named_steps["clf"], LogisticRegression):
    vec = final_model.named_steps["tfidf"]
    clf = final_model.named_steps["clf"]
    feature_names = np.array(vec.get_feature_names_out())
    coef = clf.coef_.ravel()

    top_pos_idx = np.argsort(coef)[-20:][::-1]
    top_neg_idx = np.argsort(coef)[:20]

    top_pos = pd.DataFrame({"term": feature_names[top_pos_idx], "weight": coef[top_pos_idx]})
    top_neg = pd.DataFrame({"term": feature_names[top_neg_idx], "weight": coef[top_neg_idx]})

    print("\nTop 20 fitur paling mendukung kelas Positif:")
    display(top_pos)

    print("\nTop 20 fitur paling mendukung kelas Negatif:")
    display(top_neg)

# ==========================================================
# 13) SIMPAN MODEL PIPELINE
# ==========================================================
MODEL_PATH = "uci331_sentiment_model.joblib"
joblib.dump(final_model, MODEL_PATH)
print("Model disimpan:", MODEL_PATH)

# (Opsional) unduh dari Colab
# from google.colab import files
# files.download(MODEL_PATH)

# ==========================================================
# 14) CONTOH PREDIKSI KALIMAT BARU
# ==========================================================
samples = [
    "This product is amazing and works perfectly.",
    "Terrible experience. I will never buy this again.",
    "Not bad, but could be better."
]
pred = final_model.predict(samples)

print("\nContoh prediksi kalimat baru:")
for s, yhat in zip(samples, pred):
    if hasattr(final_model.named_steps["clf"], "predict_proba"):
        score = float(final_model.predict_proba([s])[0, 1])
        print(f"- [{yhat}] (p_pos={score:.3f}) {s}")
    else:
        print(f"- [{yhat}] {s}")
```
