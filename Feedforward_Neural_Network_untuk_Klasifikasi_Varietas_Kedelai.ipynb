```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Pertemuan: 10 – Feedforward Neural Network untuk Klasifikasi Varietas Kedelai
# Dataset: Forty Soybean Cultivars from Subsequent Harvests (UCI Machine Learning Repository, ID=913)
# Repo: https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingkungan: Google Colab
# ==========================================================

# ==========================================================
# 0) SETUP: Instalasi & Import Library
# ==========================================================
!pip -q install ucimlrepo tensorflow

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["font.size"] = 11

np.random.seed(42)
tf.random.set_seed(42)

# ==========================================================
# 1) AKUISISI DATA: Ambil Dataset dari UCI (ID=913)
#    Catatan :
#    Pada dataset ini, label 'Cultivar' tersimpan di ds.data.features,
#    sehingga ds.data.targets dapat bernilai None.
# ==========================================================
ds = fetch_ucirepo(id=913)

print("Metadata ringkas (UCI):")
print(ds.metadata)

df = ds.data.features.copy()  # seluruh kolom termasuk 'Cultivar'
print("\nDaftar kolom:")
print(list(df.columns))

print("\nCuplikan data:")
display(df.head())

print("\nUkuran data:", df.shape)

# ==========================================================
# 2) VALIDASI DATA: Missing values, duplikasi, tipe data
# ==========================================================
print("\nMissing values per kolom:")
print(df.isna().sum())

dup = df.duplicated().sum()
print("\nJumlah baris duplikat:", dup)

print("\nTipe data:")
print(df.dtypes)

# ==========================================================
# 3) DEFINISI MASALAH: Klasifikasi multi-kelas (Cultivar)
#    Target : Cultivar (label varietas)
#    Fitur  : kolom lain selain Cultivar
# ==========================================================
if "Cultivar" not in df.columns:
    raise ValueError("Kolom 'Cultivar' tidak ditemukan pada dataset. Periksa struktur kolom UCI ID=913.")

y = df["Cultivar"].astype(str)
X = df.drop(columns=["Cultivar"]).copy()

# Jika ada fitur non-numerik, ubah menjadi kode kategori agar bisa diproses FFNN
non_numeric = [c for c in X.columns if not pd.api.types.is_numeric_dtype(X[c])]
if len(non_numeric) > 0:
    print("\nKolom non-numerik pada fitur:", non_numeric)
    for c in non_numeric:
        X[c] = X[c].astype("category").cat.codes

# Pastikan semua numerik dan bebas NaN
X = X.apply(pd.to_numeric, errors="coerce")
if X.isna().sum().sum() > 0:
    raise ValueError("Ada NaN pada fitur setelah konversi numerik. Periksa format nilai fitur.")

# ==========================================================
# 4) EDA RINGKAS: Statistik + Distribusi label + Korelasi fitur
# ==========================================================
print("\nRingkasan statistik fitur:")
display(X.describe().T)

class_counts = y.value_counts()

plt.figure(figsize=(10,5))
class_counts.head(15).plot(kind="bar")
plt.title("Distribusi Label Cultivar (15 Teratas)")
plt.xlabel("Cultivar")
plt.ylabel("Jumlah Sampel")
plt.tight_layout()
plt.show()

corr = X.corr(numeric_only=True)
plt.figure(figsize=(9,7))
plt.imshow(corr.values, aspect="auto")
plt.title("Heatmap Korelasi Fitur")
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.tight_layout()
plt.show()

# ==========================================================
# 5) PREPROCESSING: Encode label + Standardisasi fitur
# ==========================================================
le = LabelEncoder()
y_enc = le.fit_transform(y)
n_classes = len(le.classes_)

print("\nJumlah kelas (Cultivar):", n_classes)

# ==========================================================
# 6) PEMBAGIAN LATIH–UJI (Stratified Split)
# ==========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X.values, y_enc,
    test_size=0.20,
    random_state=42,
    stratify=y_enc
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print("\nUkuran train:", X_train_sc.shape, "| Ukuran test:", X_test_sc.shape)

# ==========================================================
# 7) MODEL: Feedforward Neural Network (FFNN)
# ==========================================================
def build_ffnn(input_dim: int, n_classes: int) -> tf.keras.Model:
    model = models.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.25),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.20),
        layers.Dense(n_classes, activation="softmax")
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

model = build_ffnn(input_dim=X_train_sc.shape[1], n_classes=n_classes)
model.summary()

cb_early = callbacks.EarlyStopping(monitor="val_loss", patience=15, restore_best_weights=True)
cb_reduce = callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=7, min_lr=1e-5)

# ==========================================================
# 8) TRAINING
# ==========================================================
history = model.fit(
    X_train_sc, y_train,
    validation_split=0.20,
    epochs=200,
    batch_size=16,
    callbacks=[cb_early, cb_reduce],
    verbose=1
)

hist = pd.DataFrame(history.history)

plt.figure(figsize=(9,4))
plt.plot(hist["loss"], label="train_loss")
plt.plot(hist["val_loss"], label="val_loss")
plt.title("Kurva Loss Training vs Validasi")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,4))
plt.plot(hist["accuracy"], label="train_acc")
plt.plot(hist["val_accuracy"], label="val_acc")
plt.title("Kurva Akurasi Training vs Validasi")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.show()

# ==========================================================
# 9) EVALUASI: Test set + Tabel metrik + Confusion matrix
# ==========================================================
test_loss, test_acc = model.evaluate(X_test_sc, y_test, verbose=0)
print(f"\nHasil Evaluasi Test -> Loss: {test_loss:.4f} | Accuracy: {test_acc:.4f}")

y_prob = model.predict(X_test_sc, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

report = classification_report(
    y_test, y_pred,
    target_names=le.classes_,
    zero_division=0,
    output_dict=True
)
report_df = pd.DataFrame(report).T

print("\nTabel ringkas metrik (accuracy, macro avg, weighted avg):")
display(report_df.loc[["accuracy", "macro avg", "weighted avg"]])

cm = confusion_matrix(y_test, y_pred)
cm_norm = cm / cm.sum(axis=1, keepdims=True)

plt.figure(figsize=(10,8))
plt.imshow(cm_norm, aspect="auto")
plt.title("Confusion Matrix (Normalisasi per Kelas)")
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.colorbar()
plt.tight_layout()
plt.show()

# ==========================================================
# 10) INFERENSI: Contoh Prediksi 3 Sampel
# ==========================================================
n_samples = 3
rng = np.random.default_rng(42)
sample_indices = rng.choice(len(X_test_sc), size=n_samples, replace=False)

X_sample = X_test_sc[sample_indices]
y_true_sample = y_test[sample_indices]

y_prob_sample = model.predict(X_sample, verbose=0)
y_pred_sample = np.argmax(y_prob_sample, axis=1)
y_conf_sample = np.max(y_prob_sample, axis=1)

rows = []
for i, idx in enumerate(sample_indices):
    rows.append({
        "Index_Test": int(idx),
        "Cultivar_Aktual": le.inverse_transform([y_true_sample[i]])[0],
        "Cultivar_Prediksi": le.inverse_transform([y_pred_sample[i]])[0],
        "Probabilitas_Maks": float(y_conf_sample[i])
    })

infer_df = pd.DataFrame(rows)
print("\nHasil inferensi 3 sampel dari test set:")
display(infer_df)

plt.figure(figsize=(7,4))
plt.bar(range(1, n_samples+1), infer_df["Probabilitas_Maks"])
plt.title("Probabilitas Maksimum Prediksi (3 Sampel)")
plt.xlabel("Sampel ke-")
plt.ylabel("Probabilitas Maks")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
