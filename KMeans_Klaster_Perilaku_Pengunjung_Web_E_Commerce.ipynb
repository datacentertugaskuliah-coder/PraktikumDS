```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Topik   : KMeans  Klaster Perilaku Pengunjung Web E-Commerce
# Dataset : Online Shoppers Purchasing Intention (UCI ID = 468)
# Sumber  : https://archive.ics.uci.edu/dataset/468/online+shoppers+purchasing+intention+dataset
# Repo    : https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingkung: Google Colab
# ==========================================================

# ----------------------------------------------------------
# 0) SETUP: Install & import library
# ----------------------------------------------------------
!pip install ucimlrepo -q

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

# Styling grafis
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

print("=== SETUP SELESAI ===")

# ----------------------------------------------------------
# 1) LOAD DATA: Ambil dataset dari UCI + cek struktur
# ----------------------------------------------------------
online = fetch_ucirepo(id=468)  # Online Shoppers Purchasing Intention (UCI)
X_features = online.data.features
y_targets = online.data.targets

# Gabungkan fitur + target (Revenue) dalam satu DataFrame
df = pd.concat([X_features, y_targets], axis=1)

print("\n=== 1) INFORMASI DATA AWAL ===")
print("Ukuran data (baris, kolom):", df.shape)
print("\n5 baris pertama:")
display(df.head())

print("\nInfo kolom:")
print(df.info())

# ----------------------------------------------------------
# 2) EDA SINGKAT: Distribusi target & ringkasan statistik
# ----------------------------------------------------------
print("\n=== 2) EDA: Distribusi Target Revenue ===")
print("Distribusi label Revenue (False = tidak beli, True = beli):")
print(df["Revenue"].value_counts(dropna=False))
print("\nProporsi Revenue:")
print(df["Revenue"].value_counts(normalize=True))

# Gambar 1. Distribusi kelas Revenue
plt.figure()
sns.countplot(x="Revenue", data=df)
plt.title("Gambar 1. Distribusi Kelas Target Revenue (Pembelian vs Tidak)")
plt.xlabel("Revenue (False = tidak beli, True = beli)")
plt.ylabel("Jumlah sesi")
plt.show()

# Ringkasan statistik fitur numerik
num_cols_all = df.select_dtypes(include=["int64", "float64"]).columns
print("\nRingkasan statistik fitur numerik:")
display(df[num_cols_all].describe().T)

# ----------------------------------------------------------
# 3) PEMILIHAN FITUR PERILAKU UNTUK KMEANS
# ----------------------------------------------------------
print("\n=== 3) PEMILIHAN FITUR PERILAKU UNTUK KMEANS ===")

behavior_cols = [
    "Administrative",
    "Administrative_Duration",
    "Informational",
    "Informational_Duration",
    "ProductRelated",
    "ProductRelated_Duration",
    "BounceRates",
    "ExitRates",
    "PageValues",
    "SpecialDay"
]

X = df[behavior_cols].copy()

print("Fitur perilaku yang digunakan:")
print(behavior_cols)

# Gambar 2. Heatmap korelasi fitur perilaku
corr = X.corr()
plt.figure(figsize=(9, 7))
sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
plt.title("Gambar 2. Korelasi Antar Fitur Perilaku Pengunjung")
plt.show()

# ----------------------------------------------------------
# 4) PREPROCESSING: Scaling fitur numerik
# ----------------------------------------------------------
print("\n=== 4) SCALING FITUR NUMERIK ===")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=behavior_cols)
print("5 baris pertama setelah scaling:")
display(X_scaled_df.head())

# ----------------------------------------------------------
# 5) MENENTUKAN JUMLAH CLUSTER k (Elbow + Silhouette)
# ----------------------------------------------------------
print("\n=== 5) PENENTUAN JUMLAH CLUSTER (k) ===")

inertias = []
sil_scores = []
k_range = range(2, 9)  # k = 2 s.d 8

for k in k_range:
    kmeans_tmp = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_tmp = kmeans_tmp.fit_predict(X_scaled)
    inertias.append(kmeans_tmp.inertia_)
    sil = silhouette_score(X_scaled, labels_tmp)
    sil_scores.append(sil)

# Gambar 3. Elbow Plot (Inertia vs k)
plt.figure()
plt.plot(k_range, inertias, marker="o")
plt.xticks(k_range)
plt.xlabel("Jumlah Cluster (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.title("Gambar 3. Elbow Plot untuk Menentukan k Optimal")
plt.grid(True)
plt.show()

# Gambar 4. Silhouette Score vs k
plt.figure()
plt.plot(k_range, sil_scores, marker="o")
plt.xticks(k_range)
plt.xlabel("Jumlah Cluster (k)")
plt.ylabel("Silhouette Score")
plt.title("Gambar 4. Silhouette Score untuk Berbagai Nilai k")
plt.grid(True)
plt.show()

# Pemilihan k secara logis: k dengan silhouette score maksimum
best_k = k_range[int(np.argmax(sil_scores))]
print(f"Silhouette Score per k: {dict(zip(k_range, np.round(sil_scores, 4)))}")
print(f"Nilai k dengan Silhouette tertinggi (dipilih): k = {best_k}")

# ----------------------------------------------------------
# 6) TRAINING KMEANS DENGAN k TERPILIH
# ----------------------------------------------------------
print("\n=== 6) TRAINING KMEANS DENGAN k TERPILIH ===")

kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=10
)

cluster_labels = kmeans.fit_predict(X_scaled)
df["cluster"] = cluster_labels

cluster_counts = df["cluster"].value_counts().sort_index()
print("Distribusi jumlah sampel per cluster:")
display(cluster_counts.to_frame(name="jumlah_sesi"))

# Tabel 1. Profil rata-rata fitur perilaku per cluster
cluster_profile = df.groupby("cluster")[behavior_cols].mean()
print("\nTabel 1. Profil Rata-Rata Fitur Perilaku per Cluster:")
display(cluster_profile)

# ----------------------------------------------------------
# 7) ANALISIS HUBUNGAN CLUSTER vs REVENUE
# ----------------------------------------------------------
print("\n=== 7) ANALISIS KLASTER vs REVENUE (INSIGHT BISNIS) ===")

# Ubah Revenue (bool) ke integer (0/1) untuk memudahkan agregasi
df["Revenue_int"] = df["Revenue"].astype(int)

cluster_revenue = df.groupby("cluster")["Revenue_int"].agg(
    jumlah_sesi="count",
    jumlah_beli="sum",
    proporsi_beli="mean"
)

print("Tabel 2. Proporsi Pembelian (Revenue=True) di Setiap Cluster:")
display(cluster_revenue)

# Gambar 5. Proporsi pembelian per cluster
plt.figure()
sns.barplot(
    x=cluster_revenue.index,
    y=cluster_revenue["proporsi_beli"]
)
plt.xlabel("Cluster")
plt.ylabel("Proporsi Pembelian (Revenue=True)")
plt.title("Gambar 5. Proporsi Pembelian per Cluster")
plt.ylim(0, 1)
plt.show()

# Gambar 6. Jumlah sesi per cluster
plt.figure()
sns.barplot(
    x=cluster_counts.index,
    y=cluster_counts.values
)
plt.xlabel("Cluster")
plt.ylabel("Jumlah Sesi")
plt.title("Gambar 6. Jumlah Sesi per Cluster")
plt.show()

# Cari cluster dengan proporsi pembelian tertinggi & terendah
best_cluster = cluster_revenue["proporsi_beli"].idxmax()
worst_cluster = cluster_revenue["proporsi_beli"].idxmin()

print(f"\nCluster dengan proporsi pembelian tertinggi  : Cluster {best_cluster}")
print(f"Cluster dengan proporsi pembelian terendah   : Cluster {worst_cluster}")

# ----------------------------------------------------------
# 8) VISUALISASI CLUSTER DALAM RUANG 2D (PCA)
# ----------------------------------------------------------
print("\n=== 8) VISUALISASI CLUSTER DENGAN PCA 2D ===")

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

df["pca1"] = X_pca[:, 0]
df["pca2"] = X_pca[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x="pca1",
    y="pca2",
    hue="cluster",
    style="Revenue",
    data=df,
    palette="Set1",
    alpha=0.7
)
plt.title("Gambar 7. Proyeksi PCA 2D  Sebaran Cluster dan Status Revenue")
plt.xlabel("Komponen Utama 1 (PCA1)")
plt.ylabel("Komponen Utama 2 (PCA2)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# 9) RINGKASAN INTERPRETASI (TEKS LOGIS)
# ----------------------------------------------------------
print("\n=== 9) RINGKASAN INTERPRETASI LOGIS ===\n")

print("1) Dataset Online Shoppers Purchasing Intention memuat sesi kunjungan e-commerce dengan")
print("   berbagai fitur perilaku, seperti jumlah halaman produk yang dikunjungi, durasi akses,")
print("   bounce rate, exit rate, PageValues, serta informasi momen khusus (SpecialDay).")
print()
print("2) Pada tahap klasterisasi, hanya fitur-fitur perilaku numerik yang digunakan sebagai input")
print("   KMeans. Langkah ini menjaga fokus pada pola browsing tanpa dipengaruhi variabel kategorikal.")
print()
print("3) Setelah semua fitur diskalakan dengan StandardScaler, jumlah cluster k ditentukan secara")
print("   logis menggunakan kombinasi Elbow Plot (Gambar 3) dan Silhouette Score (Gambar 4).")
print(f"   Nilai k dengan Silhouette tertinggi pada rentang 28 adalah k = {best_k}, sehingga")
print("   k tersebut dipakai sebagai jumlah cluster akhir.")
print()
print("4) Tabel 1 memperlihatkan profil rata-rata masing-masing cluster. Misalnya, cluster dengan")
print("   PageValues dan ProductRelated_Duration yang lebih tinggi cenderung merepresentasikan")
print("   pengunjung yang menelusuri produk lebih lama dan lebih dekat ke proses pembelian.")
print()
print("5) Revenue tidak digunakan saat pelatihan KMeans, tetapi dievaluasi kembali setelah klasterisasi.")
print("   Tabel 2, Gambar 5, dan Gambar 6 menunjukkan bahwa ada perbedaan proporsi pembelian yang cukup")
print("   jelas antar cluster. Cluster dengan proporsi pembelian tertinggi (Cluster", best_cluster, ")")
print("   dapat dianggap sebagai segmen pengunjung yang paling potensial, sedangkan cluster dengan")
print("   proporsi terendah (Cluster", worst_cluster, ") adalah segmen dengan perilaku menjauh dari transaksi.")
print()
print("6) Visualisasi PCA dua dimensi pada Gambar 7 memberikan gambaran intuitif bagaimana titik-titik")
print("   sesi pengunjung terkelompok dalam ruang fitur yang dipadatkan. Titik dengan Revenue=True tampak")
print("   cenderung berkumpul di satu atau dua cluster tertentu, menguatkan interpretasi bahwa pola")
print("   perilaku browsing memang berkaitan dengan kecenderungan melakukan pembelian.")
print()
print("Secara keseluruhan, tahapan machine learning yang dilakukan sudah mencakup:")
print("- EDA awal (struktur data, distribusi target),")
print("- pemilihan fitur relevan,")
print("- preprocessing (scaling),")
print("- penentuan jumlah cluster secara kuantitatif (Elbow + Silhouette),")
print("- training KMeans,")
print("- evaluasi berbasis insight bisnis (cluster vs Revenue),")
print("- serta visualisasi klaster dalam ruang 2D (PCA).")
print("\n=== SELESAI ===")
```
