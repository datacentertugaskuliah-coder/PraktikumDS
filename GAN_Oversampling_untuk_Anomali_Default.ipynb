```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Proyek  : Default of Credit Card Clients (UCI ID=350) â€” GAN Oversampling untuk Anomali Default
# Dataset : https://archive.ics.uci.edu/dataset/350/default+of+credit+card+clients
# Repo    : https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingk.  : Google Colab
# Tujuan  : Deteksi anomali (default=1) dengan baseline + GAN-based oversampling
# Revisi  : (1) target dipisah eksplisit untuk mencegah kebocoran fitur
#           (2) split stratified untuk imbalanced class
#           (3) evaluasi fokus PR-AUC selain ROC-AUC karena kelas minoritas
# ==========================================================

!pip -q install ucimlrepo

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report,
    roc_curve, roc_auc_score,
    precision_recall_curve, average_precision_score
)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Supaya hasil stabil
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["font.size"] = 11

# ==========================================================
# 1) AKUISISI DATA (UCI)
# ==========================================================
data = fetch_ucirepo(id=350)
X = data.data.features.copy()
y = data.data.targets.copy()

# Target di UCI biasanya bernama 'default payment next month' atau variasi serupa
# Revisi logis: pastikan y menjadi 1 kolom dan tipe int
if isinstance(y, pd.DataFrame):
    if y.shape[1] != 1:
        raise ValueError("Target memiliki lebih dari 1 kolom, cek struktur dataset.")
    y = y.iloc[:, 0]
y = y.astype(int)

print("Ukuran fitur (X):", X.shape)
print("Ukuran target (y):", y.shape)
print("Nama kolom target:", getattr(data.data.targets, "columns", ["(unknown)"]))
print("\n5 baris pertama X:")
display(X.head())
print("\nDistribusi target (0=normal, 1=anomali/default):")
display(y.value_counts().rename_axis("label").to_frame("count"))

# ==========================================================
# 2) EDA RINGKAS + VISUALISASI (Distribusi Kelas)
# ==========================================================
class_counts = y.value_counts().sort_index()
ratio_anom = class_counts.get(1, 0) / len(y)

plt.figure()
plt.bar(["Normal (0)", "Anomali/Default (1)"], [class_counts.get(0,0), class_counts.get(1,0)])
plt.title("Gambar 1. Distribusi Kelas Target (Imbalanced Check)")
plt.ylabel("Jumlah Sampel")
plt.show()

print(f"Rasio anomali (default=1): {ratio_anom:.4f} (~{ratio_anom*100:.2f}%)")

# Tabel statistik ringkas
desc = X.describe().T
desc["missing"] = X.isna().sum()
print("\nTabel 1. Statistik Ringkas Fitur (Numerik)")
display(desc[["count","mean","std","min","25%","50%","75%","max","missing"]])

# ==========================================================
# 3) PRA-PROSES: imputasi sederhana + scaling
#    Catatan integritas: dataset ini numerik; beberapa kolom bersifat kategori numerik
#    (SEX, EDUCATION, MARRIAGE). Untuk GAN sederhana, semua diperlakukan numerik,
#    lalu hasil sintetis akan di-clip/round pada kolom kategori agar tetap masuk akal.
# ==========================================================
X_proc = X.copy()

# Imputasi sederhana jika ada missing (median)
for c in X_proc.columns:
    if X_proc[c].isna().any():
        X_proc[c] = X_proc[c].fillna(X_proc[c].median())

# Split stratified (wajib untuk imbalanced)
X_train, X_test, y_train, y_test = train_test_split(
    X_proc, y, test_size=0.2, random_state=SEED, stratify=y
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print("\nUkuran train/test:", X_train_sc.shape, X_test_sc.shape)

# ==========================================================
# 4) VISUALISASI PCA 2D (untuk memahami separabilitas pola)
# ==========================================================
pca = PCA(n_components=2, random_state=SEED)
X_train_pca = pca.fit_transform(X_train_sc)

plt.figure()
plt.scatter(X_train_pca[y_train.values==0, 0], X_train_pca[y_train.values==0, 1], s=8, alpha=0.5, label="Normal (0)")
plt.scatter(X_train_pca[y_train.values==1, 0], X_train_pca[y_train.values==1, 1], s=12, alpha=0.7, label="Anomali (1)")
plt.title("Gambar 2. Proyeksi PCA 2D: Normal vs Anomali")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.show()

print("Explained variance ratio (PC1, PC2):", pca.explained_variance_ratio_)

# ==========================================================
# 5) BASELINE MODEL (Tanpa GAN): Logistic Regression + class_weight
#    Revisi logis: class_weight='balanced' untuk memperbaiki bias kelas dominan.
# ==========================================================
baseline = LogisticRegression(max_iter=2000, class_weight="balanced", random_state=SEED)
baseline.fit(X_train_sc, y_train)

proba_base = baseline.predict_proba(X_test_sc)[:, 1]
pred_base  = (proba_base >= 0.5).astype(int)

cm = confusion_matrix(y_test, pred_base)
plt.figure()
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Normal(0)","Anomali(1)"])
disp.plot(values_format="d")
plt.title("Gambar 3. Confusion Matrix Baseline (Tanpa GAN)")
plt.show()

roc_auc_base = roc_auc_score(y_test, proba_base)
pr_auc_base  = average_precision_score(y_test, proba_base)

print("Laporan Klasifikasi Baseline:\n", classification_report(y_test, pred_base, digits=4))
print(f"ROC-AUC Baseline: {roc_auc_base:.4f}")
print(f"PR-AUC  Baseline: {pr_auc_base:.4f}")

# Kurva ROC & PR
fpr, tpr, _ = roc_curve(y_test, proba_base)
prec, rec, _ = precision_recall_curve(y_test, proba_base)

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle="--")
plt.title("Gambar 4. ROC Curve Baseline")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()

plt.figure()
plt.plot(rec, prec)
plt.title("Gambar 5. Precision-Recall Curve Baseline (Imbalanced Focus)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()

# ==========================================================
# 6) GAN-BASED OVERSAMPLING (Sederhana untuk tabular)
#    Ide: latih GAN pada data anomali (y=1) agar bisa menghasilkan sampel minoritas sintetis
#    lalu gabungkan dengan train set => classifier dilatih pada data yang lebih seimbang.
#    Revisi logis:
#      - GAN dilatih hanya pada TRAIN (tidak menyentuh TEST) untuk mencegah leakage
#      - kolom kategori numerik dipulihkan dengan rounding/clip setelah inverse scaling
# ==========================================================

# Ambil data anomali dalam ruang ter-skala (lebih stabil untuk GAN)
X_anom_sc = X_train_sc[y_train.values == 1]
X_norm_sc = X_train_sc[y_train.values == 0]

print("Jumlah anomali di TRAIN:", X_anom_sc.shape[0])
print("Jumlah normal  di TRAIN:", X_norm_sc.shape[0])

input_dim = X_train_sc.shape[1]
latent_dim = 32

def make_generator(latent_dim, out_dim):
    model = keras.Sequential([
        layers.Input(shape=(latent_dim,)),
        layers.Dense(64, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(out_dim, activation="linear")  # output di ruang z-score (boleh real-valued)
    ])
    return model

def make_discriminator(in_dim):
    model = keras.Sequential([
        layers.Input(shape=(in_dim,)),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.2),
        layers.Dense(1, activation="sigmoid")
    ])
    return model

gen = make_generator(latent_dim, input_dim)
disc = make_discriminator(input_dim)

disc.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

# Model gabungan (generator -> discriminator)
disc.trainable = False
z = layers.Input(shape=(latent_dim,))
fake = gen(z)
validity = disc(fake)
gan = keras.Model(z, validity)
gan.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss="binary_crossentropy"
)

# Training loop ringkas
batch_size = 256
epochs = 2000  # bisa dikurangi jika ingin cepat
print_every = 200

d_losses, g_losses = [], []

# Jika anomali sangat sedikit, sampling dengan replacement untuk batch
anom_pool = X_anom_sc
n_anom = anom_pool.shape[0]
if n_anom < batch_size:
    print("Catatan: anomali < batch_size, batch diambil dengan replacement.")

for epoch in range(1, epochs + 1):
    # ---------------------
    # Train Discriminator
    # ---------------------
    idx = np.random.choice(n_anom, batch_size, replace=True)
    real_batch = anom_pool[idx]

    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    fake_batch = gen.predict(noise, verbose=0)

    y_real = np.ones((batch_size, 1))
    y_fake = np.zeros((batch_size, 1))

    disc.trainable = True
    d_loss_real = disc.train_on_batch(real_batch, y_real)
    d_loss_fake = disc.train_on_batch(fake_batch, y_fake)

    d_loss = 0.5 * (d_loss_real[0] + d_loss_fake[0])

    # ---------------------
    # Train Generator
    # ---------------------
    noise = np.random.normal(0, 1, (batch_size, latent_dim))
    y_gan = np.ones((batch_size, 1))  # generator ingin discriminator menganggap "real"
    disc.trainable = False
    g_loss = gan.train_on_batch(noise, y_gan)

    d_losses.append(d_loss)
    g_losses.append(float(g_loss))

    if epoch % print_every == 0:
        print(f"Epoch {epoch}/{epochs} | D_loss: {d_loss:.4f} | G_loss: {float(g_loss):.4f}")

# Plot loss
plt.figure()
plt.plot(d_losses, label="Discriminator loss")
plt.plot(g_losses, label="Generator loss")
plt.title("Gambar 6. Kurva Loss Training GAN (Oversampling Anomali)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# ==========================================================
# 7) GENERATE DATA SINTETIS + POST-PROCESS (clip/round kategori)
# ==========================================================
# Target seimbangkan train: jumlah anomali sintetis = (normal - anomali)
need = X_norm_sc.shape[0] - X_anom_sc.shape[0]
need = max(0, need)
print("Jumlah sampel sintetis anomali yang dibutuhkan:", need)

if need > 0:
    noise = np.random.normal(0, 1, (need, latent_dim))
    X_syn_anom_sc = gen.predict(noise, verbose=0)

    # Kembalikan ke ruang asli untuk pembulatan kategori, lalu scale lagi agar konsisten classifier
    X_syn_anom = scaler.inverse_transform(X_syn_anom_sc)
    X_syn_anom = pd.DataFrame(X_syn_anom, columns=X_train.columns)

    # Identifikasi kolom kategori numerik (bawaan dataset ini)
    cat_cols = [c for c in X_train.columns if c.upper() in ["SEX","EDUCATION","MARRIAGE"]]
    for c in cat_cols:
        X_syn_anom[c] = np.round(X_syn_anom[c]).astype(int)

    # Clip nilai agar masuk rentang realistis berdasar train
    for c in X_train.columns:
        lo, hi = X_train[c].min(), X_train[c].max()
        X_syn_anom[c] = X_syn_anom[c].clip(lo, hi)

    # Scale kembali
    X_syn_anom_sc2 = scaler.transform(X_syn_anom)
    y_syn = np.ones((need,), dtype=int)

    # Gabungkan training set
    X_train_gan = np.vstack([X_train_sc, X_syn_anom_sc2])
    y_train_gan = np.concatenate([y_train.values, y_syn])

else:
    X_train_gan = X_train_sc.copy()
    y_train_gan = y_train.values.copy()

print("Ukuran TRAIN setelah GAN oversampling:", X_train_gan.shape, " | prop anomali:", y_train_gan.mean())

# Tabel ringkas perubahan distribusi
tbl = pd.DataFrame({
    "Skenario": ["Train Asli", "Train Setelah GAN"],
    "Jumlah Sampel": [len(y_train), len(y_train_gan)],
    "Jumlah Anomali(1)": [int((y_train.values==1).sum()), int((y_train_gan==1).sum())],
    "Rasio Anomali": [float((y_train.values==1).mean()), float((y_train_gan==1).mean())]
})
print("\nTabel 2. Perbandingan Distribusi Train Sebelum vs Sesudah GAN")
display(tbl)

# ==========================================================
# 8) TRAIN CLASSIFIER SETELAH GAN + EVALUASI (TEST tetap asli)
# ==========================================================
clf_gan = LogisticRegression(max_iter=2000, random_state=SEED)
clf_gan.fit(X_train_gan, y_train_gan)

proba_gan = clf_gan.predict_proba(X_test_sc)[:, 1]
pred_gan  = (proba_gan >= 0.5).astype(int)

cm2 = confusion_matrix(y_test, pred_gan)
plt.figure()
disp2 = ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=["Normal(0)","Anomali(1)"])
disp2.plot(values_format="d")
plt.title("Gambar 7. Confusion Matrix Setelah GAN Oversampling")
plt.show()

roc_auc_gan = roc_auc_score(y_test, proba_gan)
pr_auc_gan  = average_precision_score(y_test, proba_gan)

print("Laporan Klasifikasi Setelah GAN:\n", classification_report(y_test, pred_gan, digits=4))
print(f"ROC-AUC Setelah GAN: {roc_auc_gan:.4f}")
print(f"PR-AUC  Setelah GAN: {pr_auc_gan:.4f}")

# Kurva ROC & PR (bandingkan)
fpr2, tpr2, _ = roc_curve(y_test, proba_gan)
prec2, rec2, _ = precision_recall_curve(y_test, proba_gan)

plt.figure()
plt.plot(fpr, tpr, label="Baseline")
plt.plot(fpr2, tpr2, label="GAN")
plt.plot([0,1],[0,1], linestyle="--")
plt.title("Gambar 8. Perbandingan ROC Curve (Baseline vs GAN)")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()

plt.figure()
plt.plot(rec, prec, label="Baseline")
plt.plot(rec2, prec2, label="GAN")
plt.title("Gambar 9. Perbandingan Precision-Recall (Baseline vs GAN)")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.show()

# Tabel metrik ringkas
metrics_tbl = pd.DataFrame({
    "Model": ["Baseline (class_weight)", "Setelah GAN oversampling"],
    "ROC-AUC": [roc_auc_base, roc_auc_gan],
    "PR-AUC": [pr_auc_base, pr_auc_gan]
})
print("\nTabel 3. Ringkasan Metrik Utama (Imbalanced Focus)")
display(metrics_tbl)

# ==========================================================
# 9) OUTPUT AKHIR: interpretasi singkat berbasis metrik
# ==========================================================
print("Catatan interpretasi:")
print("- Untuk data imbalanced, PR-AUC sering lebih informatif dibanding ROC-AUC.")
print("- Jika PR-AUC meningkat setelah GAN, berarti deteksi kelas minoritas cenderung membaik.")
print("- Jika confusions menunjukkan FP naik tajam, threshold bisa dinaikkan (mis. 0.6/0.7).")
```
