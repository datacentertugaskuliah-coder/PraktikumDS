```python
# ==========================================================
# Praktikum Data Science dan Machine Learning Berbasis Python dan UCI Repository
# Proyek  : Residential Building (UCI ID=437) â€” Prediksi Sale Prices (Regresi)
# Dataset : https://archive.ics.uci.edu/dataset/437/residential+building+data+set
# Repo    : https://github.com/datacentertugaskuliah-coder/PraktikumDS.git
# Lingk.  : Google Colab
# Model   : MLP + Dropout(0.3) + Adam(lr=1e-3) + Huber loss
# Callback: EarlyStopping + ReduceLROnPlateau
# Output  : Tabel & Gambar untuk EDA, training, evaluasi, residual
# ==========================================================

# =========================
# A) INSTALL MINIMAL (TIDAK MENGUBAH NUMPY/PANDAS/TF COLAB)
# =========================
!pip -q install ucimlrepo openpyxl

# =========================
# B) IMPORT & SEED
# =========================
import re, io, zipfile, requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

np.random.seed(42)
tf.random.set_seed(42)

plt.rcParams["figure.figsize"] = (8, 5)
plt.rcParams["font.size"] = 11

print("Versi lingkungan (ringkas):")
import sklearn
print("- sklearn:", sklearn.__version__)
print("- numpy :", np.__version__)
print("- pandas:", pd.__version__)
print("- tensorflow:", tf.__version__)

# =========================
# C) AKUISISI DATA: ucimlrepo (jika tersedia) -> fallback ZIP UCI
# =========================
def load_ucimlrepo_437():
    from ucimlrepo import fetch_ucirepo
    ds = fetch_ucirepo(id=437)
    X = ds.data.features.copy()
    y = ds.data.targets.copy()
    meta = ds.metadata
    return X, y, meta, "ucimlrepo"

def load_zip_fallback():
    zip_url = "https://archive.ics.uci.edu/static/public/437/residential%2Bbuilding%2Bdata%2Bset.zip"
    r = requests.get(zip_url, timeout=60)
    r.raise_for_status()

    z = zipfile.ZipFile(io.BytesIO(r.content))
    xlsx_files = [n for n in z.namelist() if n.lower().endswith(".xlsx")]
    if not xlsx_files:
        raise RuntimeError("File .xlsx tidak ditemukan di ZIP UCI 437.")
    xlsx_name = xlsx_files[0]

    with z.open(xlsx_name) as f:
        df = pd.read_excel(f)  # header default

    return df, "zip_fallback"

try:
    X, y, meta, source = load_ucimlrepo_437()
    print("\nSumber data:", source)
except Exception as e:
    print("\nGagal ucimlrepo, gunakan fallback ZIP. Ringkas:", str(e)[:160], "...")
    df, source = load_zip_fallback()
    print("Sumber data:", source, "| Shape:", df.shape)

    # Revisi logis: anggap 2 kolom terakhir adalah target (umum pada dataset ini)
    y = df.iloc[:, -2:].copy()
    X = df.iloc[:, :-2].copy()

print("X shape:", X.shape, "| y shape:", y.shape)
print("Kolom target (raw):", list(y.columns))

# =========================
# D) NORMALISASI NAMA TARGET + PILIH SALE PRICES SECARA ROBUST
# =========================
def clean_colname(c):
    c = str(c).strip()
    if c.lower().startswith("unnamed"):
        return None
    return c

y_cols_clean = []
for i, c in enumerate(y.columns):
    cc = clean_colname(c)
    if cc is None:
        cc = f"target_{i+1}"
    y_cols_clean.append(cc)

y.columns = y_cols_clean
print("Kolom target (clean):", list(y.columns))

def pick_sale_target(y_df: pd.DataFrame) -> str:
    cols = list(y_df.columns)

    # 1) prioritas nama kolom eksplisit
    for pat in [r"sale", r"price"]:
        hits = [c for c in cols if re.search(pat, c, re.IGNORECASE)]
        if hits:
            return hits[0]

    # 2) fallback paling konsisten untuk dataset ini: SALE PRICES = kolom terakhir
    return cols[-1]

target_col = pick_sale_target(y)
y_sale = pd.to_numeric(y[target_col], errors="coerce")

# buang NaN target agar training stabil
mask = y_sale.notna()
X = X.loc[mask].copy()
y_sale = y_sale.loc[mask].copy()

# jika ada NaN pada fitur, imputasi sederhana (median) agar pipeline tidak putus
X = X.apply(pd.to_numeric, errors="coerce")
X = X.fillna(X.median(numeric_only=True))

print("\nTarget digunakan (Sale Prices):", target_col)
print("Shape final:", X.shape, y_sale.shape)

print("\nContoh 3 baris pertama (fitur + target):")
display(pd.concat([X.head(3), y_sale.head(3).rename("sale_prices")], axis=1))

# =========================
# E) EDA (TABEL + GRAFIK)
# =========================
desc = y_sale.describe().to_frame(name="nilai").reset_index().rename(columns={"index": "statistik"})
print("\nTabel 1. Ringkasan Statistik Target Sale Prices")
display(desc)

plt.figure()
plt.hist(y_sale.values, bins=30)
plt.title("Gambar 1. Distribusi Target Sale Prices")
plt.xlabel("Sale Prices")
plt.ylabel("Frekuensi")
plt.show()

tmp = X.copy()
tmp["_target_"] = y_sale.values
corr = tmp.corr(numeric_only=True)["_target_"].drop("_target_").sort_values(key=lambda s: s.abs(), ascending=False)
top = corr.head(15)

plt.figure()
plt.bar(range(len(top)), top.values)
plt.title("Gambar 2. 15 Korelasi Numerik Teratas terhadap Sale Prices (Pearson)")
plt.xticks(range(len(top)), top.index, rotation=90)
plt.ylabel("Korelasi")
plt.tight_layout()
plt.show()

top_tbl = top.reset_index()
top_tbl.columns = ["fitur", "korelasi"]
print("\nTabel 2. Korelasi Numerik Top-15 terhadap Sale Prices")
display(top_tbl)

# =========================
# F) SPLIT + SCALING (ANTI DATA LEAKAGE)
# =========================
X_values = X.values
y_values = y_sale.values.astype(float)

X_train, X_test, y_train, y_test = train_test_split(
    X_values, y_values, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

print("\nUkuran data:")
print("Train:", X_train_sc.shape, "| Test:", X_test_sc.shape)

# =========================
# G) MODEL (MLP + DROPOUT + ADAM + HUBER) + CALLBACK
# =========================
def build_model(input_dim: int, dropout_rate: float = 0.3, lr: float = 1e-3):
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        layers.Dense(256, activation="relu"),
        layers.Dropout(dropout_rate),
        layers.Dense(128, activation="relu"),
        layers.Dropout(dropout_rate),
        layers.Dense(64, activation="relu"),
        layers.Dropout(dropout_rate),
        layers.Dense(1)
    ])
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr),
        loss=keras.losses.Huber(delta=1.0),
        metrics=[keras.metrics.MeanAbsoluteError(name="mae")]
    )
    return model

model = build_model(input_dim=X_train_sc.shape[1], dropout_rate=0.3, lr=1e-3)
print("\nRingkasan model:")
model.summary()

callbacks = [
    keras.callbacks.EarlyStopping(monitor="val_loss", patience=25, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=10, min_lr=1e-6, verbose=1),
]

# =========================
# H) TRAINING + LEARNING CURVE
# =========================
history = model.fit(
    X_train_sc, y_train,
    validation_split=0.2,
    epochs=500,
    batch_size=32,
    callbacks=callbacks,
    verbose=0
)

plt.figure()
plt.plot(history.history["loss"], label="train_loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.title("Gambar 3. Learning Curve (Huber Loss)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

plt.figure()
plt.plot(history.history["mae"], label="train_mae")
plt.plot(history.history["val_mae"], label="val_mae")
plt.title("Gambar 4. Learning Curve (MAE)")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.legend()
plt.show()

# =========================
# I) EVALUASI + VISUAL PREDIKSI & RESIDUAL
# =========================
y_pred = model.predict(X_test_sc).reshape(-1)

mae = mean_absolute_error(y_test, y_pred)
rmse = float(np.sqrt(mean_squared_error(y_test, y_pred)))
r2 = r2_score(y_test, y_pred)

metrics_df = pd.DataFrame({"metrik": ["MAE", "RMSE", "R2"], "nilai": [mae, rmse, r2]})
print("\nTabel 3. Evaluasi Model pada Data Uji")
display(metrics_df)

plt.figure()
plt.scatter(y_test, y_pred)
plt.title("Gambar 5. Aktual vs Prediksi (Sale Prices)")
plt.xlabel("Aktual")
plt.ylabel("Prediksi")
plt.show()

residual = y_test - y_pred

plt.figure()
plt.hist(residual, bins=30)
plt.title("Gambar 6. Distribusi Residual (Aktual - Prediksi)")
plt.xlabel("Residual")
plt.ylabel("Frekuensi")
plt.show()

plt.figure()
plt.scatter(y_pred, residual)
plt.title("Gambar 7. Residual Plot terhadap Prediksi")
plt.xlabel("Prediksi")
plt.ylabel("Residual")
plt.axhline(0)
plt.show()

out_df = pd.DataFrame({
    "sale_prices_aktual": y_test,
    "sale_prices_prediksi": y_pred,
    "residual": residual
}).head(10)

print("\nTabel 4. Contoh 10 Prediksi Pertama")
display(out_df)

print("\nKesimpulan:")
print(f"- MLP + Dropout(0.3) + Huber loss menghasilkan MAE={mae:.4f}, RMSE={rmse:.4f}, R2={r2:.4f} pada data uji.")
print("- EarlyStopping mengembalikan bobot terbaik berbasis val_loss, ReduceLROnPlateau menstabilkan optimasi saat val_loss stagnan.")
```
